## Задание

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш

Выгодна ли игра игроку?

Сколько в среднем приобретает
или теряет игрок за одну игру?
```
777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
*77: 5 руб.
**7: 3 руб.
*00: 2 руб.
**0: 1 руб.

* – любая цифра
```

#### Дополнительное задание повышенной сложности. 
Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

### Решение

````python
import numpy as np
import re

# Теоретическая вероятность
# Всего 1000 вариантов цифр
# Выигрышные из них: 5 целых + 10*2 из *77 и *00 + 100*2 из **7 и **0, итого 225
# Т.е. теоретическая вероятность выигрыша в принципе 225/1000 или 22,5%
# Среднее значение = 0.845 и считается как сумма произведений результата (выигрыша) на его вероятность для каждой комбинации.
# Учитывая, что 1 рубль тратится на попытку, теоретический результат попытки будет = -0.155, а значит убыточен
x = (200+100+50+15+10+10*5+90*3+10*2+90)/1000
theory_res = x-1


pdict = {
'777': 200 
,'999': 100 
,'555': 50 
,'333': 15 
,'111': 10 
,'\d{1,1}77': 5
,'\d{2,2}7': 3 
,'\d{1,1}00': 2 
,'\d{2,2}0': 1 
}


i = 0

res = []

while i <= 1000000:
# Поскольку некоторые числа вроде 100 или 777 будут подходить под разные маски, будем брать самую "выгодную"
    wp = []
    account = -1
    num = str(np.random.randint(0, 1000)).rjust(3,'0')
    for k, v in pdict.items():
        if re.match(k,num):
            wp.append(v)
    if len(wp) > 0:
        wp = max(wp)
        account += wp
    i+=1
    res.append(account)

res = sum(res) / len(res)


print(f'''Теоретический выигрыш/проигрыш: {theory_res}
По методу Монте-Карло: {res}
Разница: {theory_res-res}''')

# Однозначно вывод - невыгодно
````
