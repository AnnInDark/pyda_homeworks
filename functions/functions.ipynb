### Решение (сразу задания 1 и 2 чтобы не дублировать входные данные и часть функций)

````python
## Исходные данные ##

# перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
# перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

## Функции ##

def ShelfList():
    """
    Возвращает список номеров полок
    """
    return 'Текущий перечень полок: ' +  ', '.join(list(directories.keys()))

def DocumentList():
    """
    Возвращает список документов "красивым выводом"
    """
    res = 'Текущий список документов:\n'
    for i in documents:
       dnum = i['number']
       dtype = i['type']
       down = i['name']
       shelf = ShelfByDocument(dnum).split()[-1]
       res += f'№: {dnum}, тип: {dtype}, владелец: {down}, полка хранения: {shelf} \n'
    return res

def ShelfByDocument(docnum):
    """
    Возвращает номер полки по номеру документа
    """
    for n, d in directories.items():
        if d.count(docnum) > 0:
            return f'Документ хранится на полке: {n}'
    return 'Документ не найден в базе'

def DocOwnerByNum(docnum):
    """
    Возвращает владельца документа по его номеру
    """
    for i in documents:
        if i['number'] == str(docnum):
            return 'Владелец документа: ' + i['name']
    return 'Документ не найден в базе'

def ShelfAdd(snum):
    """
    Ф-ция добавления полки
    """
    if list(directories.keys()).count(snum) == 0:
        directories[snum] = []
        return 'Полка добавлена. '
    else:
        return 'Такая полка уже существует. '

def ShelfDel(snum):
    """
    Ф-ция удаления полки
    """
    if list(directories.keys()).count(snum) == 0:
        return 'Такой полки не существует. '
    elif directories[snum]:
        return 'На полке есть документ(ы), удалите их перед удалением полки. '
    else:
        directories.pop(snum)
        return 'Полка удалена. '

## Функции для доп. (необязательной) части ##

def DocAdd(*params):
    """
    Ф-ция добавления документа
    """
    if list(directories.keys()).count(params[3]) == 0:
        return 'Такой полки не существует. Добавьте полку командой as.\n'
    else:
        NewDoc = {}
        NewDoc['number'] = params[0]
        NewDoc['type'] = params[1]
        NewDoc['name'] = params[2]
        documents.append(NewDoc)
        directories[params[3]].append(params[0])
        return 'Документ добавлен. \n'

def DocDel(DocNum):
    """
    Ф-ция удаления документа
    """
    # удаляем с полки
    for v in directories.values():
        if v.count(DocNum) > 0:
            v.remove(DocNum)
    # удаляем сам документ
    for i, item in enumerate(documents):
        if item['number'] == DocNum:
            documents.pop(i)
            return 'Документ удален. \n'
    return 'Документ не найден в базе. \n'

def DocMove(DocNum, Shelf):
    """
    Ф-ция перемещения документа между полками.
    Сначала проверяем наличие документа и новой полки.
    Далее т.к. update всегда = delete + insert удаляем документ со старой полки и записываем в новую
    """
    # сначала проверки
    if [i['number'] for i in documents].count(DocNum) == 0:
        return 'Документ не найден в базе.\n' + DocumentList()
    if list(directories.keys()).count(Shelf) == 0:
        return 'Такой полки не существует. ' + ShelfList()
    for k, v in directories.items():
    # удаляем с полки
        if v.count(DocNum) > 0:
            v.remove(DocNum)
    # добавляем на полку
        if k == Shelf:
            print('add')
            v.append(DocNum)
    return 'Документ перемещен. \n' + DocumentList()



## Основная исполнительная часть ##

command = input('Введите команду: ')

### MAIN PART ###
if command == 'p': # узнать владельца документа по его номеру
    print(DocOwnerByNum(input('Введите номер документа: ')))
elif command == 's': # по номеру документа узнать на какой полке он хранится
    print(ShelfByDocument(input('Введите номер документа: ')))
elif command == 'l': # полную информацию по всем документам
    print(DocumentList())
elif command == 'ads': # добавить новую полку
    print(ShelfAdd(input('Введите номер полки: ')), ShelfList())
elif command == 'ds': # удалить существующую полку из данных (только если она пустая)
    print(ShelfDel(input('Введите номер полки: ')), ShelfList())
### EXTENDED ###
elif command == 'ad': # добавить новый документ в данные
    print(DocAdd(
        input('Введите номер документа: '),
        input('Введите тип документа: '),
        input('Введите владельца документа: '),
        input('Введите полку для хранения: ')), DocumentList())
elif command == 'd': # удалить документ из данных
    print(DocDel(input('Введите номер документа: ')), DocumentList())
elif command == 'm': # переместить документ с полки на полку
    print(DocMove(
        input('Введите номер документа: '),
        input('Введите номер полки: ')))
````
